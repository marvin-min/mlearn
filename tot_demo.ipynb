```python
import os
from openai import OpenAI
from dotenv import load_dotenv

from langchain_demo2_debug import response

load_dotenv()
```

```python
client = OpenAI(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)
```

```python
prompt = """
小明100米跑步成绩：10.5秒，1500米跑步成绩：3分20秒，铅球成绩：12米。他适合参加那些搏击运动训练？
请根据以上成绩，分析候选人在速度、耐力、力量三方面素质的分档。分档包括：强（3），中（2），弱（1）三挡
需要速度强的运动有哪些。请给出10个例子，需要耐力强的运动有哪些，给出10个例子，需要力量最强的运动有哪些，给出10个例子
分别分析上面的10个运动对速度、耐力、力量方面的素质的要求：强（3），中（2），弱（1）
根据上面分析：生成一篇小明适合那种运动训练的分析报告
"""
```

```python
def get_completion(prompt, model="qwen-max-latest"):
  messages = [
    {"role":"user", "content":prompt},
  ]
  response = client.chat.completions.create(
    model=model,
    messages=messages,
  )
  return response.choices[0].message.content
```

```python
print(get_completion(prompt))
```
