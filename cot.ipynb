```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)
```

```python
#qwen2.5-coder-3b-instruct qwen-max-latest
def get_completion(prompt, model="qwen2.5-coder-3b-instruct"):
  messages = [{"role": "user", "content": prompt}]
  response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0,
  )
  return response.choices[0].message.content
```

```python
cot_prompt ="""
Q：我去超市买了8个苹果，吃了2个，又买了3个，我总共有几个苹果？
A: 刚开始有8个苹果，吃了2个，苹果数是8-2=6，又买了3个，6+3=9
Q：我去超市买了3个梨，吃了1个，又买了5个，之后又吃了2个，我总共有几个梨？
A:
Q：我去超市买了3个梨，吃了1个，又买了5个苹果，之后又吃了2个，我总共有几个梨？
A:
"""


rs = get_completion(cot_prompt)
print(rs)
```
